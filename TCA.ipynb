```python
import pandas as pd
import matplotlib.pyplot as plt

# Create percentage table
data = pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize="index") * 100

# Plot stacked bar chart
ax = data.plot(kind="bar",
               stacked=True,
               figsize=(4,4),
               color=["lightblue", "lightgreen"])

# Add percentage labels
for container in ax.containers:
    labels = [f"{v.get_height():.1f}%" if v.get_height() > 0 else "" for v in container]
    ax.bar_label(container, labels=labels, label_type="center", color="black")

plt.title("Churn Percentage by Senior Citizen")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.show()
```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

plt.figure(figsize=(12, 15))

for i, col in enumerate(columns, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=df, x=col, hue='Churn')
    plt.title(col)
    plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
#These count plots provide an overview of the distribution of customers across different telecom services. Basic services such as Phone Service and Internet Service are widely used, while additional services like Online Security, Online Backup, Device Protection, Tech Support, Streaming TV, and Streaming Movies show varying levels of customer adoption. These visualizations help identify the most and least popular services and provide a foundation for further churn analysis.
```
